```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Train a simple perceptron to realize a logical function OR
```

```python
# make some training data
Xs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Ys = np.array([[0], [1], [1], [1]])
```

```python
# initialize the weight and bias with small random values
weights = np.random.rand(2, 1) * 0.1 - 0.05
bias = np.random.rand(1) * 0.1 - 0.05
print(weights)
print(bias)
```

```python
# set the limitation of max iteration
T = 20
# set the value of learning rate
lr = 0.25

cnt = 0
for i in range(T):
    flag = 1
    for j in range(Xs.shape[0]):
        X = Xs[j]
        Y = Ys[j]

        Y_hat = np.where((np.dot(X.T, weights) + bias) > 0, 1, 0)
        
        cnt += 1
        print(' * Round: %d'%cnt)
        print('X1 = %d'%X[0])
        print('X2 = %d'%X[1])
        print('Y_hat = %d'%Y_hat)
        print("Y_target = %d"%Y)
        
        if Y_hat == Y:
            continue
        
        # weight <-- weight - learning_rate * (y_hat - y_target) * x
        weights -= np.dot(X.reshape(2,1), Y_hat-Y).reshape(2,1)
        # bias <-- bias - learning_rate * (y_hat - y_target) 
        bias -= lr *  Y_hat-Y
        flag = 0
    if flag:
        print('Finish.')
        break
    else:
        print('Start a new round.')
```

```python
x1cord1 = []
x2cord1 = []
x1cord0 = []
x2cord0 = []
for i in range(Xs.shape[0]):
    if Ys[i] == 1:
        x1cord1.append(Xs[i, 0])
        x2cord1.append(Xs[i, 1])
    else:
        x1cord0.append(Xs[i, 0])
        x2cord0.append(Xs[i, 1])

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x1cord1, x2cord1, s = 30, c='green')
ax.scatter(x1cord0, x2cord0, s = 30, c='red', marker='s')
x_fig = np.arange(-0.2, 1.0, 0.1)
y_fig = (-bias[0] - weights[0] * x_fig) / weights[1]
ax.plot(x_fig, y_fig)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
